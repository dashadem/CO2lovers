```python
from mpmath import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [0.73658, 3.22225e-2, 0.28157e-2, 0.20421e-2, 0.79333e-3, 0.58367e-3, 0.27394e-3, 0.17645e-3, 0.64422e-4, 0.33006e-4]

b_0 = 0.28924
omega_ksi = []
for ksi in np.linspace(0.0,1,100):
    omega = 0
    for i in range(10):
        omega += a[i]*(1-ksi**2)**(1/3)*gegenbauer(2*(i+1) -2, 5/6, ksi)
    omega+=b_0*(1-ksi)
    omega_ksi.append(omega)
```

```python
x = np.linspace(0.0, 1, 100)

# Построение графика
plt.plot(x,omega_ksi)  
plt.xlabel('ksi')  
plt.ylabel(' omega')  
plt.title('omega when gamma_0 = 0.86998') 
plt.grid(True)  
plt.show()  
```

```python


# Построение графика
plt.plot(-np.array(omega_ksi)/2, x) 
plt.plot(np.array(omega_ksi)/2, x)
plt.xlabel('tipo x')  
plt.ylabel('ksi')  
plt.title('omega when gamma_0 = 0.86998')  
plt.grid(True)  
plt.show()  
```
